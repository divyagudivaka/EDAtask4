```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("train.csv")
df.head()
```

```python
sns.set(style="whitegrid")
df.info()
df.describe()
df.isnull().sum()
```

```python
sns.countplot(x="Survived", data=df)
plt.show()
```

The majority of passengers did not survive. Survivors are significantly fewer compared to non-survivors, indicating that the Titanic disaster had a high fatality rate.


```python
sns.histplot(df["Age"], kde=True)
plt.show()
```

  Most passengers were between 2040 years old. There are fewer passengers in the older age groups, and some missing values in the Age column.


```python
sns.histplot(df["Fare"], kde=True)
plt.show()
```

Most fares are concentrated between 0 and 100, with a few very high outliers. Higher fares often indicate first-class tickets.



```python
sns.countplot(x="Sex", hue="Survived", data=df)
plt.show()
```

Females had a much higher survival rate compared to males. This supports the women and children first rescue policy.


```python
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.show()
```

First-class passengers had the highest survival rate, followed by second class. Third-class passengers had the lowest survival rate.


```python
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df)
plt.show()
```

Passengers who paid higher fares (often first-class) had a better chance of survival. Younger passengers also show slightly better survival rates.


```python
sns.heatmap(df.select_dtypes(include=["number"]).corr(), annot=True, cmap="coolwarm")
plt.show()
```

Survival is positively correlated with Fare and negatively correlated with Passenger Class (lower class number means higher class). Age shows a weaker correlation with survival.


Summary 
- The majority of passengers did not survive the Titanic disaster.
- Females had a much higher survival rate than males.
- First-class passengers survived more often than those in lower classes.
- Younger passengers had slightly higher chances of survival.
- Higher fares were generally associated with better survival odds.
- The Age column has missing values that might require imputation in further analysis.
- Fare distribution shows a few extreme outliers.
- Correlation analysis confirms the importance of class and fare in survival chances.
